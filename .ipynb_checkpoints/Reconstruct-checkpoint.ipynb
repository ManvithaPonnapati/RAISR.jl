```julia
using Distributions
INTERPOLATED_FOLDER = "/Users/manvithaponnapati/RAISR/train_data/interpolated/"
low_res_im = load(INTERPOLATED_FOLDER*"15011.jpg")

channel_low_res_im = channelview(low_res_im)

channel_low_res_im_r = channel_low_res_im[1,:,:]
channel_low_res_im_r_fl = convert(Array{Float64},channel_low_res_im_r)


channel_low_res_im_g = channel_low_res_im[2,:,:]
channel_low_res_im_g_fl = convert(Array{Float64},channel_low_res_im_g)



channel_low_res_im_b = channel_low_res_im[3,:,:]
channel_low_res_im_b_fl = convert(Array{Float64},channel_low_res_im_b)


scale_factor = 2
Qangle = 24
Qstrength = 3
Qcoherence = 3
patchsize = 3

start_index = Int(ceil(patchsize/2))
weights_matrix = rand(Normal(0, 2), patchsize*patchsize)
weights_matrix = Array(Diagonal(weights_matrix))

colorview_rgb = convert(Array{Float64},channelview(low_res_im))
color_ve = 0.21*colorview_rgb[1,:,:]+ 0.72*colorview_rgb[2,:,:]+0.07*colorview_rgb[3,:,:]

w,h = size(color_ve)
predictHR = zeros(w,h)
#dummy img
random_img = channelview(low_res_im) 
random_img = convert(Array{Float64},random_img)
for row in start_index:Int(w)-start_index
    for col in start_index:Int(h)-start_index
                patch = color_ve[row-1:row+1,col-1:col+1]
                gx,gy = imgradients(patch)
                gx = reshape(gx,patchsize*patchsize,1)
                gy = reshape(gy,patchsize*patchsize,1)
                GT = transpose([gx gy])
                GTWG = GT*weights_matrix*transpose(GT)
                eigen_max = eigmax(GTWG)
                eigen_min = eigmin(GTWG)
                eigen_vector_max = eigvecs(GTWG)[:,1]
                eigen_vector_min = eigvecs(GTWG)[:,2]
                gradient_angle = atan2(eigen_vector_max[2],eigen_vector_max[1])
                if gradient_angle < 0
                    gradient_angle = gradient_angle + pi
                end
                lamda = abs(sqrt(complex(eigen_max)))/Qstrength
                u =(sqrt(complex(eigen_max)) -  sqrt(complex(eigen_min)))/(sqrt(complex(eigen_max)) +  sqrt(complex(eigen_min)))/Qcoherence
                angle = floor(gradient_angle/pi*Qangle)


                if lamda < 0.0001
                    strength = 1
                elseif lamda > 0.001
                    strength = 3
                else
                    strength = 2
                end

                u = abs(u)
                if u < 0.25
                    coherence = 1
                elseif u > 0.5
                    coherence = 3
                else
                    coherence = 2
                end


                # Bound the output to the desired ranges
                if angle > 23
                    angle = 23
                elseif angle <= 0
                    angle = 1
                end
                angle = Int(angle)
                # Get pixel type
                pixeltype = ((row-start_index) % scale_factor) * scale_factor + ((col-start_index) % scale_factor)
                patch_ravel = reshape(patch,patchsize*patchsize)
        
                patch_r = channel_low_res_im_r_fl[row-1:row+1,col-1:col+1]
                patch_r = convert(Array{Float64},patch_r)
                patch_r_ravel = reshape(patch_r,patchsize*patchsize)
                p_r_im = dot(patch_r_ravel,hfilters[angle,strength,coherence,pixeltype+1,:])
                random_img[1,row,col]=p_r_im
        
                patch_g = channel_low_res_im_g_fl[row-1:row+1,col-1:col+1]
                patch_g = convert(Array{Float64},patch_g)
                patch_g_ravel = reshape(patch_g,patchsize*patchsize)
                p_g_im = dot(patch_g_ravel,hfilters[angle,strength,coherence,pixeltype+1,:])
                random_img[2,row,col]=p_g_im
        
                println(hfilters[angle,strength,coherence,pixeltype+1,:])
                patch_b = channel_low_res_im_b_fl[row-1:row+1,col-1:col+1]
                patch_b = convert(Array{Float64},patch_b)
                patch_b_ravel = reshape(patch_b,patchsize*patchsize)
                p_b_im = dot(patch_b_ravel,hfilters[angle,strength,coherence,pixeltype+1,:])
                random_img[3,row,col]=p_b_im
        
    end
end
```
