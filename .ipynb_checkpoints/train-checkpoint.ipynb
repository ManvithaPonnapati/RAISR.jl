```julia
using Images
using RandomMatrices
using Distributions

#Scaling factor set to 2 for initial testing
#That means P1-P4 type of "patches" - Number of pizel types = scaling factor ^ 2
scaling_factor = 2 #Scaling factor determines the number of pixel types/filters
patchsize = 11
patchmargin = Int(floor(patchsize/2))

n = 9 #For calculating the type of pixel
gradientmargin = Int(sqrt(n))

Qangle = 24
Qstrength = 3
Qcoherence = 3

weighting = Diagonal(rand(Float64,n,n) * (2 - (-1)) - 1)
weighting = Array(weighting)

Q_angle = 24
Q_strength = 3
Q_coherence = 3

# Matrix preprocessing
# Preprocessing normalized Gaussian matrix W for hashkey calculation
GROUND_TRUTH_FOLDER = "/Users/manvithaponnapati/RAISR/train_data/ground_truth/"
LOW_RES_FOLDER = "/Users/manvithaponnapati/RAISR/train_data/low_res/"
INTERPOLATED_FOLDER = "/Users/manvithaponnapati/RAISR/train_data/interpolated/"

Q = zeros((Qangle, Qstrength, Qcoherence, scaling_factor*scaling_factor, patchsize*patchsize, patchsize*patchsize))
V = zeros((Qangle, Qstrength, Qcoherence, scaling_factor*scaling_factor, patchsize*patchsize))
H = zeros((Qangle, Qstrength, Qcoherence, scaling_factor*scaling_factor, patchsize*patchsize))
mark = zeros((Qstrength*Qcoherence, Qangle, scaling_factor*scaling_factor))
```

```julia
#Function to calculate the hash table key for a given pixel
function hashkey(x,w)
    
    # Calculate gradient
    # Transform 2D matrix into 1D array
    gx,gy = imgradients(x)
    gx = reshape(gx,3*3,1)
    gy = reshape(gy,3*3,1)
    G = transpose(hcat(gx,gy))
    GTWG = G*w*transpose(G)
    largest_eign = eigvecs(GTWG) 
    min_eig = eigmin(GTWG)
    max_eig = eigmax(GTWG)
    strength = abs(sqrt(complex(max_eig)))
    coherence = (sqrt(complex(max_eig)) -  sqrt(complex(min_eig)))/(sqrt(complex(max_eig)) +  sqrt(complex(min_eig)))
    theta = atan2(largest_eign[2,1],largest_eign[1,1])
    
    #Gradient Anglae
    if theta < 0
        theta = theta + pi
    end
    angle = floor(theta/pi*Q_angle)
    if angle < 12
        anglek = 1
    elseif angle > 12
        anglek = 3
    else
        anglek = 2
    end
    
    #COHERENCE
    q_coh = abs(coherence)/Q_coherence
    
    if q_coh < 0.25
        coherence = 1
    elseif q_coh > 0.5
        coherence = 3
    else
        coherence = 2
    end
    
    #STRENGTH
    q_str = strength/Q_strength
    if q_str < 0.0001
        strength = 1
    elseif q_str > 0.001
        strength = 3
    else
        strength = 2
    end  
    
    return anglek,strength,coherence
end
```

```julia
#for row in Int(margin)+1: Int(w-margin)-1
#             for col in Int(margin)+1:Int(h-margin)-1
filenames = readdir(INTERPOLATED_FOLDER)
margin = floor(max(patchsize,n)/2)
yi_real_patches = []
patch_type = []
count = 1
for file in filenames
    if (contains("$file", ".jpg")) == true
        yi_real = load(INTERPOLATED_FOLDER*file)
        yi = convert(Array{Float64},Gray.(load(INTERPOLATED_FOLDER*file)))
        xi = convert(Array{Float64},Gray.(load(GROUND_TRUTH_FOLDER*file))) 
        Ajt = Float64[]
        bjt = Float64[]
        w,h = size(yi)
        println("Image $count")
        count = count+1
        for row in Int(margin+1):Int(w-margin)-1
            for col in Int(margin+1):Int(h-margin)-1
                println("$row,$col")
                patch = yi[row-5:row+5, col-5:col+5]
                patch_real= yi_real[row-5:row+5, col-5:col+5]
                # Get gradient block
                push!(yi_real_patches,patch_real)
                gradientblock = yi[row-1:row+1, col-1:col+1]
                # Calculate hashkey
                gradient,strength,coherence = hashkey(gradientblock,weighting)
#                 println("Gradient : $gradient , Strength: $strength,Coherence: $coherence")
                #Unravel the patch
                patch_unravel = reshape(patch,size(patch)[1]*size(patch)[1],1)
                Ajt = [Ajt;transpose(patch_unravel)]
                bjt = [bjt;xi[row,col]] 
                pixeltype = ((row-Int(margin)) % scaling_factor) * scaling_factor + ((col-Int(margin)) % scaling_factor)
                push!(patch_type,pixeltype)
               
                ATA = transpose(Ajt)*Ajt
                ATb = transpose(Ajt)*bjt
                Q[gradient,strength,coherence,pixeltype+1,:,:] += ATA
                V[gradient,strength,coherence,pixeltype+1,:,:] += ATb
                
            end
        end
    end
end
```

```julia
# Conjugate Gradients Solver
function cgls(A, b)
    height, width = size(A)
    x = zeros((height))
    while(true)
        sumA = sum(A)
        
        if (sumA < 100)
            break
        end
        if (det(A) < 1)
            A = A + eye(height, width) * sumA * 0.000000005
        else
            x = inv(A)*b
            break
        end
    end
    return x
end
        

for i in 1:24
    for j in 1:3
        for k in 1:3
            for m in 1:4
                Qtype = (Q[i,j,k,m,:,:])
                Vtype = (V[i,j,k,m,:])
                H[i,j,k,m,:] = cgls(Qtype,Vtype)
            end
        end
    end
end
```
