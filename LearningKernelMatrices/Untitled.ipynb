# Initialize three random guess for kernel matrics 
1. Polynomial Kernel K1
2. Gaussian Kernel K2
3. Linear Kernel K3
Find the optimal coefficients  mhu1, mhu2, mhu3 to obtain the optimal kernel for the given dataset

```julia
function linear_kernel(x1, x2)
    return dot(x1, x2)
end

function polynomial_kernel(x, y, p=3)
    return (1 + dot(x, y)) ^ p
end

function gaussian_kernel(x, y, sigma=5.0)
    return exp(-norm(x-y)^2 / (2 * (sigma ^ 2)))
end
```

```julia
Pkg.add("LIBSVM")
```

```julia
Pkg.add("RDatasets")
```

```julia
using RDatasets, LIBSVM
```

```julia
# Load Fisher's classic iris data
iris = dataset("datasets", "iris")

# LIBSVM handles multi-class data automatically using a one-against-one strategy
labels = convert(Vector, iris[:Species])

# First dimension of input data is features; second is instances
instances = convert(Array, iris[:, 1:4])'
feature_length, isntance_length = size(instances)
K1 = zeros(isntance_length,isntance_length)
K2 = zeros(isntance_length,isntance_length)
K3 = zeros(isntance_length,isntance_length)
for i = 1:isntance_length
    for j = 1:isntance_length
        K1[i,j]=linear_kernel(instances[:,i],instances[:,j])
        K2[i,j]=polynomial_kernel(instances[:,i],instances[:,j])
        K3[i,j]=gaussian_kernel(instances[:,i],instances[:,j])
    end
end
```

```julia
Pkg.add("Mosek")
```

```julia
Pkg.checkout("Mosek","b0.7")
```

```julia
Pkg.build("Mosek")
```

```julia
K = [K1 K2 K3]
# Solving Equation (41)
using JuMP
using Mosek

m = Model(solver=MosekSolver())
#let's define all the possible varibales required for the model

@variable(m, t)
@variable(m, C)
@variable(m, e)
@variable(m, lambda)
@variable(m,mhu[1:3])
@constraint(m,sum(sum(K[i]*mhu[i] for i=1:3))== 1)
@constraint(m,sum(K[i]*mhu[i] for i=1:3)>= 0)
@constraint(m,[mhu[1]*K[1]+mhu[2]*K[2]+mhu[3]*K[3]])
@variable(m,V>=0)
```

```julia
# Train SVM on half of the data using default parameters. See documentation
# of svmtrain for options
model = svmtrain(instances[:, 1:2:end], labels[1:2:end]);

# Test model on the other half of the data.
(predicted_labels, decision_values) = svmpredict(model, instances[:, 2:2:end]);

# Compute accuracy
@printf "Accuracy: %.2f%%\n" mean((predicted_labels .== labels[2:2:end]))*100
```
